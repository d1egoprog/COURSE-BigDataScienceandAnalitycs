```python
greet = "Bienvenido a Python y a Tecnicas de NLP."
print(greet)
```

```python
import pandas as pd

df = pd.read_csv('export-2020-7-23-16-9-54.dsv', encoding='utf-8', sep=r'\|\|\|', engine='python')
print(df)
```

```python
year = df['year'][3]
type(year)
```

```python
dfStatuses = df['status']
print(dfStatuses)
type(dfStatuses)
```

```python
s = ' '.join(dfStatuses)
print(s)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = set(stopwords.words('spanish')) 
print(stop)
```

```python
import re
s = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', s)
print(s)
```

```python
from nltk.tokenize import word_tokenize 
nltk.download('punkt')
tokenGreet = word_tokenize(greet)
print(tokenGreet)
```

```python
import nltk
from nltk.tokenize import word_tokenize 
nltk.download('punkt')

tokenGreet = word_tokenize(s)
tokenGreetwithOut = []
#also - tokenGreetwithOut = [word for word in tokenGreet if word.isalnum()]
for word in tokenGreet:
    if word.isalnum() and word != 'RT' and word not in stop:
        tokenGreetwithOut.append(word)
print(tokenGreetwithOut)
```

```python
from nltk.tokenize import RegexpTokenizer
nltk.download('punkt')

# Good Regex - Reference https://www3.ntu.edu.sg/home/ehchua/programming/howto/Regexe.html#zz-1.1
tokenizer = RegexpTokenizer(r'\w+')
tokenGreet = tokenizer.tokenize(greet)

print(tokenGreet)
```

```python
from nltk.tokenize import word_tokenize 
nltk.download('punkt')

tokens = tokenizer.tokenize(s)
withoutStop = [] 

#also - withoutStop = [word for word in tokens if w not in stop]
for w in tokens: 
    if w not in stop: 
        withoutStop.append(w) 
        
print(withoutStop)
```

```python
withoutStop = [] 
#also - withoutStop = [word for word in tokens if w not in stop  and w != 'RT']
for w in tokens: 
    if w not in stop and w != 'RT' and w != '@': 
        withoutStop.append(w) 
        
print(withoutStop)
```

```python
from nltk.tokenize import word_tokenize 
nltk.download('punkt')
tokens = tokenizer.tokenize(s)
withoutStop = [word for word in tokens if word not in stop and word != 'RT']
print(withoutStop)
```

```python
bof = pd.Series(tokenGreetwithOut).value_counts()
print(bof)
```

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("spanish")
stemmed_spanish = [stemmer.stem(item) for item in tokenGreetwithOut]
print(stemmed_spanish)
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')
text = ','.join(tokenGreetwithOut)
wordcloud.generate(text)
wordcloud.to_image()
```

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# Helper function
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))

number_topics = 6
number_words = 15

listofstatuses = df['status'].to_numpy()
count_vectorizer = CountVectorizer(stop_words=stop)
listofConversations = count_vectorizer.fit_transform(listofstatuses)

lda = LDA(n_components=number_topics)
lda.fit(listofConversations)

print("Topics found via LDA:")
print_topics(lda, count_vectorizer, number_words)
```
